```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,10)
y = x**2
plt.plot(x,y)
plt.title('y = x²')
plt.show()
```

```python
dados={'Nome':['Ana','Bruno','Carlos'],'Salario':[3000,5000,4000]}
df=pd.DataFrame(dados)
print(df)
```

```python
plt.bar(df['Nome'],df['Salario'])
plt.title('Salários')
plt.show()
```
